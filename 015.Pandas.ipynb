#**Series**

A Series is very similar to a NumPy array. What differentiates the NumPy array from a Series, is that a Series can have axis labels, meaning it can be indexed by a label, instead of just a number location. It also doesn't need to hold numeric data, it can hold any arbitrary Python Object.

```python
import numpy as np
import pandas as pd
```

```python
labels=[ 'a', 'b', 'c']
mydata=[10,20,30]
arr=np.array(mydata)
d={'a':10, 'b':20, 'c':30}
```

```python
pd.Series(data=mydata)
```

```python
pd.Series(data=mydata,index=labels)
```

```python
pd.Series(arr) #using NumPy Arrays
```

```python
pd.Series(arr,labels) #using NumPy Arrays
```

```python
pd.Series(d) #using dictionaries
```

```python
#A pandas Series can hold a variety of object types:
pd.Series(data=labels)
```

```python
#The key to using a Series is understanding its index. Pandas makes use of these index names or numbers by allowing for fast look ups of information (works like a hash table or dictionary).
ser1=pd.Series([1,2,5,3],index=['USA','Germany','USSR','Japan'])
```

```python
ser1
```

```python
ser1['USA']
```

```python
ser3=pd.Series(data=labels)
ser3
```

```python
ser3[0]
```

```python
ser2=pd.Series([1,2,5,4],index=['USA','Germany','Italy','Japan'])
ser2
```

```python
ser1+ser2 #add index.
```

#**Merging, Joining, and Concatenating**

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],      ##we specifically indicate index otherwise it automatically gives 0,1,2,3
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='key')
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on=['key1', 'key2']) #only rows matching
```

```python
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='right', on=['key1', 'key2'])
#In a right join, all the rows from the right DataFrame are included in the result, and only the matching rows from the left DataFrame are included.
#If there are no matches, NaN values are used for the missing values from the left DataFrame.
```

```python
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

#**Joining**
Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame.

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left
```

```python
right
```

```python
left.join(right)
```

```python
right.join(left)
```

```python
left.join(right, how='outer')
```
