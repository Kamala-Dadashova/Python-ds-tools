## **Dynamic Typing**

Python uses dynamic typing, meaning you can reassign variables to different data types. This makes Python very flexible in assigning data types; it differs from other languages that are statically typed.

```python
languages=2
languages
```

```python
languages=['Englihs', 'German']
languages
```

# **Formatting a string**

```python
print("Her name is {}".format("Kamala D"))
```

```python
print("{2} {0} {1}".format("fox", "dog", "chicken"))
```

```python
print("The {g} {f} {h}".format(f="love", g="gratitude", h="smile"))
```

```python
result=100/777
result
```

```python
print("The result is {r}". format(r=result)) #value:width.precision f
```

```python
print("The result is {r:1.3f}". format(r=result))
```

```python
print("The result is {r:10.3f}". format(r=result))
```

```python
print("The result is {r:1.6f}". format(r=result))
```

```python
#Another way of doing that
name="Kamala"
print(f"Hello, her name is {name}")
```

```python
print('{0:8} | {1:9}'.format('Fruit', 'Quantity'))
print('{0:8} | {1:9}'.format('Apples', 3.))
print('{0:8} | {1:9}'.format('Oranges', 10))
```

```python
print('{0:<8} | {1:^8} | {2:>8}'.format('Left','Center','Right'))
print('{0:<8} | {1:^8} | {2:>8}'.format(11,22,33))
```
